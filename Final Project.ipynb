#Analyzing Canada's COVID-19 Data

#Data Source

The data is downloaded from [European Centre for Disease Prevention and Control](https://www.ecdc.europa.eu/en/publications-data/download-todays-data-geographic-distribution-covid-19-cases-worldwide) 

```python
%matplotlib inline
import pandas
```

```python
df = pandas.read_excel('s3://chaz-ia241/covid_data.xls')
Canada_data = df.loc[ df['countriesAndTerritories']=='Canada']
```

#Small view of Canada's COVID-19 Data

#Number of cases in Canada in November

```python
Canada_data.sum()['cases']
```

```python
print("The number of Covid-19 cases in Canada as of November 2020 is 460,743 cases.")
```

```python
sum_cases_per_day = Canada_data.groupby('month').sum()['cases']
sum_cases_per_day.plot()
```

#Deaths related to number of cases

```python
Canada_data.plot.scatter(x='cases',y='deaths')
```

#Month with highest number of cases

```python
month_Canada_cases = df.groupby('month').sum()['cases']
month_Canada_cases.plot.bar()
month_Canada_cases.idxmax()
```
